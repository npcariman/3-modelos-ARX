## Preprocesamiento de los datos del Equivital

```python
# Abrimos la base de datos
mydb = mysql.connector.connect(
    host='localhost',
    user='root',
    password='7461143',
    database='datos_brutos'
)

# Extraemos la informacion en un dataframe
df_eq = pd.read_sql("SELECT * FROM equivital_5", mydb)   # Cargamos todos los datos 
df_eq.drop('id', axis=1, inplace=True)                   # Eliminamos el indice
df_eq.set_index('datetime', inplace=True)                # Definimos datetime como indice
df_eq.sort_index(inplace=True)                           # Ordenamos en base a datetime
# Mostramos los resultados
print('Tamano de la tabla: {} filas y {} columnas'.format(df_eq.shape[0], df_eq.shape[1]))
df_eq.head(5)
```

### Preprocesamos los datos del HR y los unidos al dataframe

```python
# Preprocesamiento de los datos antes de submuestrear
# Definimos las columnas que vamos a extraer
confidence_th = 95               # Parametro de intervalo de confianza a utilizar
u_HR = df_eq.loc[df_eq['HR_Confidence'] > confidence_th, 'HR']    # Exraemos las series
u_HR = u_HR[u_HR > 30]                    # Adicionalmente se pide que HR sea mayor que 30

# Realizamos un proceso de resampling con regla de agrupacion la media
u_HR = u_HR.resample('5T').mean()

# Cambiamos el nombre y lo agregamos al dataframe original
u_HR.name = 'u_HR'

# Creamos un dataframe para ir guardando los datos
df_eq_2 = copy.copy(df)
df_eq_2.rename(columns={'sensor_glucose':'y'}, inplace=True)

# Agregamos los datos
df_eq_2 = df_eq_2.join(u_HR)
df_eq_2.head(3)
```

## Preprocesamiento de los datos del Holter

Lo primero es cargar la base de datos

```python
# Abrimos la base de datos
mydb = mysql.connector.connect(
    host='localhost',
    user='root',
    password='7461143',
    database='datos_ordenados'
)

# Extraemos la informacion en un dataframe
df_holter = pd.read_sql("SELECT * FROM holter_ordenados", mydb)   # Cargamos todos los datos 
df_holter.drop('id', axis=1, inplace=True)                        # Eliminamos el indice
df_holter.set_index('datetime', inplace=True)                     # Definimos datetime como indice
df_holter.sort_index(inplace=True)                                # Ordenamos en base a datetime
# Mostramos los resultados
print('Tamano de la tabla: {} filas y {} columnas'.format(df_holter.shape[0], df_holter.shape[1]))
df_holter.head(5)
```

Dado que la tasa de muestreo de este sensor es de 15 minutos, se realiza un resampling de 5 minutos, donde los datos faltantes son rellenados con `nan` (Para que coincida con los datos del sensor de glucosa). Luego se realiza una interpolación de llenado hacia adelante (`metod=pad` en pandas) para llenar los datos de modo correcto.

```python
# Extraemos los datos del holter a una variable
u_holter = copy.copy(df_holter[['Sys', 'Dia', 'HR', 'Mean']])

# Realizamos un proceso de resampling con regla de agrupacion la media
u_holter = u_holter.resample('5T').mean()

# Realizamos la interpolacion para llenar los espacios libres producto del resampling 
u_holter = u_holter.interpolate(method='pad', limit=15)

# Agregamos los ultimos valores del holter
ultima_fecha = u_holter.index
ultima_fecha = pd.to_datetime(ultima_fecha[-1].value)
ultimos_val = u_holter.iloc[-1,:]
u_holter.loc[ultima_fecha + pd.Timedelta(minutes=5)] = ultimos_val
u_holter.loc[ultima_fecha + pd.Timedelta(minutes=10)] = ultimos_val

# Cambiamos el nombre
u_holter.rename(columns={'Sys':'u_sys', 'Dia':'u_dia', 'HR':'u_HR', 'Mean':'u_mean'}, inplace=True)

u_holter.head(4)
```

Agregamos la salida al dataframe

```python
# Creamos un dataframe con la variable de salida
df_holter = copy.copy(df)
df_holter.rename(columns={'sensor_glucose':'y'}, inplace=True)

# Realizamos una union con los datos
df_holter = df_holter.join(u_holter)

# Calculamos el porcentaje de datos disponibles respecto al original
N_total = len(df_holter)
df_holter_2 = df_holter.dropna()
print('Datos del Holter: {:0.2f} %'.format(100* len(df_holter_2)/N))
print('Numero de datos: {}'.format(len(df_holter_2)))
df_holter_2.head(5)
```
