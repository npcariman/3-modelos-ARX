# 4 - Modelo ARX

## 4.1 Descripción

El cuarto modelo a utilizar es un modelo autoregresivo con entrada exógena, $\text{ARX}(n_a, n_b)$, definido como:

\begin{equation}
y(t) + a_1 y(t-1) + \dots + a_{n_a}y(t-n_a) = b_1 u(t-1) + \dots + b_{n_b} u(t - n_b) + e(t)
\end{equation}

donde $n_a$ es el orden de la regresión y $n_b$ es el orden de la entrada exógena.

```python
# Importamos las librerias necesarias para trabajar
from statsmodels.regression.linear_model import yule_walker
from statsmodels.graphics.tsaplots import plot_acf
from sklearn.metrics import mean_squared_error
from numpy.fft import fft, fftfreq, fftshift
from mpl_toolkits.mplot3d import Axes3D
from scipy import signal
from matplotlib import cm
from datetime import time
from matplotlib import rc
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mysql.connector
import seaborn as sns
import pandas as pd
import numpy as np
import datetime
import copy
import sklearn

# Seteamos el estilo de los graficos
sns.set(style="whitegrid")

# Configuramos los graficos con latex
plt.rc('text', usetex=True)

# Funcion para la metrica de la trayectoria
def error_trayectoria(vector):
    '''
    Se calcula una norma de error como la diferencia entre la prediccion y la medicion
    real para un tiempo t. Se utilizara la norma euclidiana
    '''
    y_predict = vector['y+1':'y+6']
    y_real = vector['y']
    error = (y_predict - y_real) ** 2
    error = sum(error) / len(error)
    return np.sqrt(error)

def phi_X(R_X, gamma, Ts=5 * 60):
    arg_max = R_X.argmax()
    R_X_wind = R_X[arg_max - gamma: arg_max + gamma + 1]
    wind = np.hanning(len(R_X_wind))
    phi_X = fft(R_X_wind * wind)
    freq = fftfreq(len(phi_X), Ts)
    phi_X = pd.Series(phi_X, index=freq)
    phi_X = phi_X[freq > 0]
    return phi_X
```

## 4.4 Modelo óptimo para meal

Utilizando MATLAB, se buscó el óptimo y se extrajeron los parámetros. Estos se guardaron en un archivo `.csv`. Se utilizó la convención en la documentación de MATLAB para crear la función.

```python
# Orden del modelo
n_a = 48
n_b = 6

# Valor de los vectores del polinomio A(q) y B(q)
A = np.genfromtxt('A_meal.csv', delimiter=',')
B = np.genfromtxt('B_meal.csv', delimiter=',')

# Generamos el modelo
def arx_meal_model(z):
    global A, B
    x = z[-len(A) + 1:]
    phi_A = np.flip(-A[1:])
    u = z[:len(B)]

    phi_B = np.flip(B)
    sum_A = np.inner(phi_A, x)
    sum_B = np.inner(phi_B, u)
    return sum_A + sum_B
```

Con el modelo entrenado, comenzamos a analizar el modelo

```python
# pasos a predecir
k = 6

######################################
# Dataframe para predecir
######################################
u_train = copy.copy(data_train['u_meal'])
Z_train = pd.DataFrame(copy.copy(data_train['y']))
u_test = copy.copy(data_test['u_meal'])
Z_test = pd.DataFrame(copy.copy(data_test['y']))

# Agregamos la variable endogena
for i in range(n_a - 1):
    Z_train['y-{}'.format(i+1)] = Z_train['y'].shift(i+1)
    Z_test['y-{}'.format(i+1)] = Z_test['y'].shift(i+1)

# Agregamos la variable exogena
Z_train['u_meal'] = u_train
Z_test['u_meal'] = u_test
for i in range(n_b):
    Z_train['u_meal-{}'.format(i+1)] = u_train.shift(i+1)
    Z_test['u_meal-{}'.format(i+1)] = u_test.shift(i+1)
Z_train = Z_train[Z_train.columns[::-1]]
Z_test = Z_test[Z_test.columns[::-1]]

######################################
# Prediccion
######################################
# Realizamos la prediccion
for i in range(k):
    Z_train['y+{}'.format(i+1)] = Z_train.apply(arx_meal_model, axis=1)
    Z_test['y+{}'.format(i+1)] = Z_test.apply(arx_meal_model, axis=1)
    
# Se debe desfasar la prediccion para que coincida con el valor el tiempo   
for i in range(k):
    Z_train['y+{}'.format(i+1)] = Z_train['y+{}'.format(i+1)].shift(i+1)
    Z_test['y+{}'.format(i+1)] = Z_test['y+{}'.format(i+1)].shift(i+1)

######################################
# Calculo de los errores
######################################

# 1 paso adelante (5 minutos)
Z_train['e_1_paso'] = Z_train['y+1'] - Z_train['y']
Z_test['e_1_paso'] = Z_test['y+1'] - Z_test['y']

# 6 pasos adelante (30 minutos)
Z_train['e_6_paso'] = Z_train['y+6'] - Z_train['y']
Z_test['e_6_paso'] = Z_test['y+6'] - Z_test['y']

# Trayectoria (5 a 30 minutos)
Z_train['e_trajec'] = Z_train.apply(error_trayectoria, axis=1) 
Z_test['e_trajec'] = Z_test.apply(error_trayectoria, axis=1)
```

### Modelo optimo para bolo de insulina

```python
# Orden del modelo
n_a = 48
n_b = 4

# Valor de los vectores del polinomio A(q) y B(q)
A = np.genfromtxt('A_bolo.csv', delimiter=',')
B = np.genfromtxt('B_bolo.csv', delimiter=',')

# Generamos el modelo
def arx_meal_model(z):
    global A, B
    x = z[-len(A) + 1:]
    phi_A = np.flip(-A[1:])
    u = z[:len(B)]

    phi_B = np.flip(B)
    sum_A = np.inner(phi_A, x)
    sum_B = np.inner(phi_B, u)
    return sum_A + sum_B
```

Con el modelo entrenado, comenzamos a analizaar el modelo

```python
# pasos a predecir
k = 6

######################################
# Dataframe para predecir
######################################
u_train = copy.copy(data_train['u_bolo'])
Z_train = pd.DataFrame(copy.copy(data_train['y']))
u_test = copy.copy(data_test['u_bolo'])
Z_test = pd.DataFrame(copy.copy(data_test['y']))

# Agregamos la variable endogena
for i in range(n_a - 1):
    Z_train['y-{}'.format(i+1)] = Z_train['y'].shift(i+1)
    Z_test['y-{}'.format(i+1)] = Z_test['y'].shift(i+1)

# Agregamos la variable exogena
Z_train['u_bolo'] = u_train
Z_test['u_bolo'] = u_test
for i in range(n_b):
    Z_train['u_bolo-{}'.format(i+1)] = u_train.shift(i+1)
    Z_test['u_bolo-{}'.format(i+1)] = u_test.shift(i+1)
Z_train = Z_train[Z_train.columns[::-1]]
Z_test = Z_test[Z_test.columns[::-1]]

######################################
# Prediccion
######################################
# Realizamos la prediccion
for i in range(k):
    Z_train['y+{}'.format(i+1)] = Z_train.apply(arx_meal_model, axis=1)
    Z_test['y+{}'.format(i+1)] = Z_test.apply(arx_meal_model, axis=1)
    
# Se debe desfasar la prediccion para que coincida con el valor el tiempo   
for i in range(k):
    Z_train['y+{}'.format(i+1)] = Z_train['y+{}'.format(i+1)].shift(i+1)
    Z_test['y+{}'.format(i+1)] = Z_test['y+{}'.format(i+1)].shift(i+1)

######################################
# Calculo de los errores
######################################

# 1 paso adelante (5 minutos)
Z_train['e_1_paso'] = Z_train['y+1'] - Z_train['y']
Z_test['e_1_paso'] = Z_test['y+1'] - Z_test['y']

# 6 pasos adelante (30 minutos)
Z_train['e_6_paso'] = Z_train['y+6'] - Z_train['y']
Z_test['e_6_paso'] = Z_test['y+6'] - Z_test['y']

# Trayectoria (5 a 30 minutos)
Z_train['e_trajec'] = Z_train.apply(error_trayectoria, axis=1) 
Z_test['e_trajec'] = Z_test.apply(error_trayectoria, axis=1)
```
